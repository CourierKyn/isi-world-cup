```python
import pandas as pd
import math
import numpy as np
import re
data=pd.read_excel("赛题1数据材料\赛题1数据集\年报-的对外提供保证担保信息.xlsx")
data
```

```python
for i in range(len(data)):
    if (type(data.iloc[i,1])==np.str and  data.iloc[i,1]!="企业选择不公示"):
        if(data.iloc[i,1][-2:]=="万元"):
            data.iloc[i,1]=float(data.iloc[i,1][:-2])*10000
        else:data.iloc[i,1]=float(data.iloc[i,1])
data
```

```python
data['主债权种类'].value_counts()
```

```python
df1=data.set_index("保证的期间").loc[["期限","企业选择不公示","限期","未约定"],:].reset_index().set_index("企业编号")
dummy_1=pd.get_dummies(df1["保证的期间"])
dummy_1=dummy_1.groupby("企业编号").sum().T
dummy_1
```

```python
df2=data.set_index("保证的方式").loc[["连带保证","企业选择不公示","一般保证","未约定"],:].reset_index().set_index("企业编号")
dummy_2=pd.get_dummies(df2["保证的方式"])
dummy_2=dummy_2.groupby("企业编号").sum().T
dummy_2
```

```python
df3=data.set_index("主债权种类").loc[["合同","企业选择不公示","其他"],:].reset_index().set_index("企业编号")
dummy_3=pd.get_dummies(df3["主债权种类"])
dummy_3=dummy_3.groupby("企业编号").sum().T
dummy_3
```

```python
result_12=dummy_1.append(dummy_2).append(dummy_3).fillna(0)#三表连接
result_12
```
