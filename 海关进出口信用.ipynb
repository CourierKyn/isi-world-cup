```python
import pandas as pd
data=pd.read_excel("赛题1数据材料\赛题1数据集\海关进出口信用.xlsx")
data
```

```python
df1=data.set_index("经济区划").loc[["一般经济区域","高新技术产业开发区","经济特区","经济技术开发全区"],:].reset_index().set_index("企业编号")
df1
```

```python
dummy_1=pd.get_dummies(df1["经济区划"])
dummy_1=dummy_1.groupby("企业编号").sum().T
dummy_1
```

```python
df2=data.set_index("年报情况").loc[["已报送","未报送","超期未报送","超期报送","不需要"],:].reset_index().set_index("企业编号")
df2
```

```python
dummy_2=pd.get_dummies(df2["年报情况"])
dummy_2=dummy_2.groupby("企业编号").sum().T
dummy_2
```
