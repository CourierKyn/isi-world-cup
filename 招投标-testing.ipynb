```python
import pandas as pd
import numpy as np
import re
import datetime
import dateutil
```

```python
data=pd.read_excel('赛题1测试数据集/招投标.xlsx',index_col=0)
data
```

```python
data.drop_duplicates(keep='first',inplace=True)
```

```python
count=pd.DataFrame(data.index.value_counts()).T.reindex(columns=list(range(4001,4526))).fillna(0)
count.index=pd.Index(['招投标_count'])
```

```python
data_1=pd.get_dummies(data['公告类型'],prefix='公告类型').groupby('企业编号').sum().T.reindex(columns=list(range(4001,4526))).fillna(0)
```

```python
data_2=pd.get_dummies(data['中标或招标'],prefix='中标或招标').groupby('企业编号').sum().T.reindex(columns=list(range(4001,4526))).fillna(0)
```

```python
#计算时间间隔
def to_datetime(cell):
    if not isinstance(cell, str):
        return cell
    try:
        return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
    except:
        print(cell)
```

```python
data_3=data[['发布时间']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
```

```python
data=pd.concat([data_1,data_2,data_3,count])
```

```python
fillindex=['公告类型_中标', '公告类型_公开招标', '公告类型_其他', '公告类型_其它', '公告类型_单一', '公告类型_变更',
       '公告类型_合同', '公告类型_废标', '公告类型_成交', '公告类型_招标', '公告类型_拟建', '公告类型_流标',
       '公告类型_竞争性谈判', '公告类型_竞价', '公告类型_竞谈', '公告类型_结果变更', '公告类型_询价', '公告类型_违规',
       '公告类型_预告', '公告类型_验收', '中标或招标_中标', '中标或招标_招标', '发布时间', '招投标_count']
```

```python
data=data.reindex(fillindex).fillna(0)
```

```python
data.to_pickle('招投标-testing.pickle')
```
