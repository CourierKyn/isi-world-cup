```python
import pandas as pd
import numpy as np
import re
import datetime
import dateutil
```

```python
data=pd.read_excel('赛题1测试数据集/作品著作权.xlsx',index_col=0)
data
```

```python
#去除同一公司信息相同的行
data.drop_duplicates(keep='first',inplace=True)
```

```python
count=pd.DataFrame(data.index.value_counts()).T.reindex(columns=list(range(4001,4526))).fillna(0)
count.index=pd.Index(['作品著作权_count'])
```

```python
typelist=['美术','文字','F 美术','类似摄制电影方法创作的作品','其他','工程设计图、产品设计图','L 其他作品']
data=data[data['作品著作权类别'].isin(typelist)]
```

```python
data
```

```python
data_1=pd.get_dummies(data['作品著作权类别'],prefix='作品著作权类别').groupby('企业编号').sum().T.reindex(columns=list(range(4001,4526))).fillna(0)
```

```python
#处理时间间隔
def to_datetime(cell):
    if not isinstance(cell, str):
        return cell
    try:
        return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
    except:
        print(cell)
```

```python
data['作品著作权登记日期'].value_counts()
```

```python
data.replace(['国作登字-2012-A-00066200','国作登字-2013-F-00084814','国作登字-2013-F-00083953','国作登字-2012-A-00066200','国作登字-2013-F-00084814','国作登字-2013-F-00083953'],np.nan,inplace=True)
```

```python
data_2=data[['作品著作权登记日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
```

```python
data_3=data[['作品著作权创作完成日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
```

```python
data['作品著作权首次发布日期'].replace('42285',np.nan,inplace=True)
```

```python
data_4=data[['作品著作权首次发布日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
```

```python
data=pd.concat([data_1,data_2,data_3,data_4,count])
```

```python
filldata=['作品著作权类别_F 美术', '作品著作权类别_L 其他作品', '作品著作权类别_其他', '作品著作权类别_工程设计图、产品设计图',
       '作品著作权类别_文字', '作品著作权类别_类似摄制电影方法创作的作品', '作品著作权类别_美术', '作品著作权登记日期',
       '作品著作权创作完成日期', '作品著作权首次发布日期', '作品著作权_count']
```

```python
data=data.reindex(filldata)
data.loc['作品著作权类别_L 其他作品',:]=data.loc['作品著作权类别_L 其他作品',:].fillna(0)
```
