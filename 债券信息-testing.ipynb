```python
import pandas as pd
import numpy as np
import re
import datetime
import dateutil
```

```python
data=pd.read_excel('赛题1测试数据集/债券信息.xlsx',index_col=0)
data
```

```python
data.columns
```

```python
data.drop_duplicates(keep='first',inplace=True)
count=pd.DataFrame(data.index.value_counts()).T.reindex(columns=list(range(4001,4526))).fillna(0)
count.index=pd.Index(['招投标_count'])
data
```

```python
data['债券信用评级']='AAA'
data
```

```python
data_1=pd.get_dummies(data['债券信用评级'],prefix='债券信用评级').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_1.loc[:,:]=0
fillindex=['债券信用评级_AA', '债券信用评级_AA+', '债券信用评级_AA-', '债券信用评级_AAA', '债券信用评级_B',
       '债券信用评级_C', '债券信用评级_CC']
data_1=data_1.reindex(fillindex).fillna(0)
data_1
```

```python
data_2=pd.get_dummies(data['债券品种'],prefix='债券品种').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526))).fillna(0)
fillindex=['债券品种_中央企业债', '债券品种_企业债券', '债券品种_公司、企业债',
       '债券品种_地方企业债', '债券品种_沪企债', '债券品种_深企债', '债券品种_银行间企债']
data_2=data_2.reindex(fillindex).fillna(0)
data_2
```

```python
#计算时间间隔
def to_datetime(cell):
    if not isinstance(cell, str):
        return cell
    try:
        return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
    except:
        print(cell)
```

```python
data_3=data[['发行日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_3
```

```python
#适用于虚拟变量用sum时
def limit(cell):
    if cell > 1:
        return 1
    return cell
```

```python
data_4=data[['兑付日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_4
```

```python
data_5=data[['计划发行总额（亿元）']].groupby('企业编号').sum().T.reindex(columns=list(range(4001,4526)))
data_5
```

```python
data['流通场所']='银行间债券市场'
```

```python
data_6=pd.get_dummies(data['流通场所'],prefix='流通场所').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_6.loc[:,:]=0
fillindex=[ '流通场所_上交所', '流通场所_其他', '流通场所_银行间债券市场',
       '流通场所_银行间债券市场 柜台', '流通场所_银行间债券市场 深交所']
data_6=data_6.reindex(fillindex).fillna(0)
data_6
```

```python
data_7=data[['票面利率（%）']].groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_7
```

```python
data_8=pd.get_dummies(data['付息方式'],prefix='付息方式').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526))).fillna(0)
fillindex=['付息方式_到期一次还本付息',
       '付息方式_半年付息', '付息方式_按季付息', '付息方式_按年付息', '付息方式_附息式固定利率', '付息方式_附息式浮动利率']
data_8=data_8.reindex(fillindex).fillna(0)
data_8
```

```python
data=pd.concat([data_1,data_2,data_3,data_3,data_3,data_4,data_5,data_6,data_7,data_8,count])
data
```
