```python
import pandas as pd
data=pd.read_excel("赛题1数据材料\赛题1数据集\纳税A级年份.xlsx")
data
```

```python
data=data.drop_duplicates(subset=['企业编号','纳税A级年份'],keep='first')  #去除重复的年份，重复的年份为垃圾数据
data
```

```python
data["纳税A级年数"]=1
df=data.drop(columns="纳税A级年份") #统计纳税A级年份的次数，不考虑具体哪一年
df
```

```python
result_11=df.groupby(["企业编号"]).sum().T
result_11
```
