```python
df = pd.read_excel('赛题1测试数据集/项目信息.xlsx', index_col='企业编号')
df['企业规模'] = df['企业规模'].str.replace(
    '500-1000人', '500-2000人'
).replace(
    '1000人以上', '500-2000人'
).replace(
    '16228', '10000人以上'
).replace(
    '40507', '10000人以上'
).replace(
    '289', '150-500人'
).replace(
    '300-500人', '150-500人'
).replace(
    '10594', '10000人以上'
).replace(
    '17478', '10000人以上'
).replace(
    '3926', '2000-5000人'
).replace(
    '少于15人', '15人以下'
).replace(
    '589', '500-2000人'
)
df
```

```python
def limit(cell):
    if cell > 1:
        return 1
    return cell
```

```python
industry = pd.get_dummies(df['行业'], prefix='行业').groupby('企业编号').sum().T
industry = industry.applymap(limit).reindex(columns=list(range(4001, 4526))).fillna(0)
industry
```

```python
business_scale = pd.get_dummies(df['企业规模'], prefix='企业规模').groupby('企业编号').mean().T
business_scale = business_scale.reindex(columns=list(range(4001, 4526))).fillna(0)
business_scale
```

```python
operational_status = pd.get_dummies(df['运营状态'], prefix='运营状态').groupby('企业编号').mean().T
operational_status = operational_status.reindex(columns=list(range(4001, 4526))).fillna(0)
operational_status
```

```python
def to_day(cell):
    if not isinstance(cell, str):
        return cell
    return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
```

```python
project_establishment_time = df[['项目成立时间']].applymap(to_day).groupby('企业编号').mean().T
project_establishment_time = project_establishment_time.reindex(columns=list(range(4001, 4526)))
project_establishment_time
```

```python
data=pd.concat([industry, business_scale, operational_status, project_establishment_time])
```

```python
fillindex=['行业_人工智能', '行业_企业服务', '行业_光电', '行业_公用事业', '行业_农业', '行业_化工', '行业_医疗健康',
       '行业_地产建筑', '行业_房产家居', '行业_文娱传媒', '行业_旅游', '行业_材料', '行业_汽车交通', '行业_消费生活',
       '行业_物流', '行业_环保', '行业_生产制造', '行业_电商', '行业_硬件', '行业_能源矿产', '行业_金融',
       '企业规模_10000人以上', '企业规模_150-500人', '企业规模_15人以下', '企业规模_2000-5000人',
       '企业规模_50-150人', '企业规模_500-2000人', '企业规模_5000-10000人', '运营状态_融资中',
       '运营状态_运营中', '项目成立时间']
```

```python
data=data.reindex(fillindex)
data.iloc[:-1,:]=data.iloc[:-1,:].fillna(0)
data.to_pickle('项目信息-testing.pickle')
```
