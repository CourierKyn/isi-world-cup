```python
from IPython.display import display
import pandas as pd
import csv
import random
import re
# import pyperclip
import numpy as np
import seaborn as sns
import requests
import matplotlib, matplotlib.pyplot as plt
import pprint
import os
import dateutil
import datetime

data=pd.read_excel("赛题1测试数据集/年报-社保信息.xlsx").drop_duplicates(keep='first').replace(
    "企业选择不公示",np.nan).replace("选择不公示",np.nan)
                                                                 
data
```

```python
def change(cell):
    return float(cell)
data["年报年份"]=data[["年报年份"]].applymap(change)
data
```

```python
data=data.set_index('年报年份').loc[[2016.0,2017.0],:].reset_index()
data
```

```python
def get_number(cell):
    if not isinstance(cell, str):
        return cell
    _ = re.findall(r'[\d.]+',cell)
    if not _:
        return None
    return float(_[0])
df=data.applymap(get_number).set_index(["企业编号","年报年份"])
df
```

```python
result=df.groupby(["企业编号","年报年份"]).mean().unstack().T
result
```
