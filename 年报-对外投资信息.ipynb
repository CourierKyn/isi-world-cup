```python
df = pd.read_excel('赛题1数据材料/赛题1数据集/年报-对外投资信息.xlsx', index_col='企业编号')
df
```

```python
investment_amount = df[['投资金额']].groupby('企业编号').sum().T
investment_amount = investment_amount.reindex(columns=list(range(1001, 4001)))
investment_amount
```

```python
investment_ratio = df[['投资占比']].groupby('企业编号').mean().T
investment_ratio = investment_ratio.reindex(columns=list(range(1001, 4001)))
investment_ratio
```

```python
pd.concat([investment_amount, investment_ratio]).to_pickle('')
```
