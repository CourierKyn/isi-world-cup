```python
from IPython.display import display
import pandas as pd
import csv
import random
import re
# import pyperclip
import numpy as np
import seaborn as sns
import requests
import matplotlib, matplotlib.pyplot as plt
import pprint
import os
import dateutil
import datetime

data=pd.read_excel("赛题1测试数据集/年报-企业基本信息.xlsx",index_col="企业编号")
data
```

```python
def change(cell):
    cell=float(cell)
    return cell

data["年报年份"]=data[["年报年份"]].applymap(change)
data
```

```python
def limit(cell):
    if cell > 1:
        return 1
    return cell
```

```python
data=data.reset_index()
data=data.set_index(['企业编号','年报年份'])
data
```

```python
dummy_2=pd.get_dummies(data["是否有网站或网点"],prefix='是否有网站或网点')
dummy_2=dummy_2.groupby(["企业编号",'年报年份']).mean()
dummy_2["是否有网站或网点_否"]=dummy_2["是否有网站或网点_否"]
dummy_2["是否有网站或网点_是"]=dummy_2["是否有网站或网点_是"]+dummy_2["是否有网站或网点_有"]
dummy_2=dummy_2.drop(columns=["是否有网站或网点_有"]).unstack().T
dummy_2=dummy_2.applymap(limit).reindex(columns=list(range(4001, 4526))).fillna(0)
dummy_2=dummy_2.reindex(pd.MultiIndex.from_product([['是否有网站或网点_否', '是否有网站或网点_是'], [2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0, 2016.0, 2017.0, 2018.0]])).fillna(0)
dummy_2
```

```python
dummy_5=pd.get_dummies(data["是否提供对外担保"],prefix='是否提供对外担保')
dummy_5=dummy_5.groupby(["企业编号","年报年份"]).mean().unstack().T
dummy_5=dummy_5.applymap(limit).reindex(columns=list(range(4001, 4526))).fillna(0)
dummy_5=dummy_5.reindex(pd.MultiIndex.from_product([['是否提供对外担保_否', '是否提供对外担保_是'], [2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0, 2016.0, 2017.0, 2018.0]])).fillna(0)
dummy_5
```

```python
def to_day(cell):
    if not isinstance(cell, str):
        return cell
    return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
dd = data[['发布日期']].applymap(to_day).groupby('企业编号').mean().T
dd = dd.reindex(columns=list(range(4001, 4526)))
dd
```

```python
dn=data.replace("企业选择不公示",np.nan)
dn
```

```python
def get_number(cell):
    if not isinstance(cell, str):
        return cell
    _ = re.findall(r'[\d.]+',cell)
    if not _:
        return None
    return float(_[0])
```

```python
dn=dn[["从业人数"]].applymap(get_number).groupby(["企业编号","年报年份"]).mean().unstack().T
dn=dn.reindex(columns=list(range(4001, 4526)))
dn=dn.reindex(pd.MultiIndex.from_product([['从业人数'], [2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0, 2016.0, 2017.0, 2018.0]])).fillna(0)
dn
```
