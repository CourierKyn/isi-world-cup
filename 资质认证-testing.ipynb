```python
import pandas as pd
import numpy as np
import re
import datetime
import dateutil
```

```python
data=pd.read_excel('赛题1测试数据集/资质认证.xlsx',index_col=0)
data
```

```python
#去除完全相同的行
data.drop_duplicates(keep='first',inplace=True)
```

```python
count=pd.DataFrame(data.index.value_counts()).T.reindex(columns=list(range(4001,4526))).fillna(0)
count.index=pd.Index(['资质认证_count'])
```

```python
typelist=['CCC产品认证证书','其他自愿性工业产品认证','CCC/家用和类似用途设备','CCC/机动车辆及安全附件','CCC/音视频设备']
data=data[data['证书名称'].isin(typelist)]
```

```python
data.columns
```

```python
data_1=pd.get_dummies(data['证书名称'],prefix='证书名称').groupby('企业编号').sum().T.reindex(columns=list(range(4001,4526))).fillna(0)
```

```python
#处理时间间隔
def to_datetime(cell):
    if not isinstance(cell, str):
        return cell
    try:
        return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
    except:
        print(cell)
```

```python
data_2=data[['有效期起止日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
```

```python
data_3=data[['有效期截至日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
```

```python
data_4=pd.get_dummies(data['状态'],prefix='状态').groupby('企业编号').sum().T.reindex(columns=list(range(4001,4526))).fillna(0)
```

```python
data=pd.concat([data_1,data_2,data_3,data_4,count])
```

```python
data
```

```python
fillindex=['证书名称_CCC/家用和类似用途设备', '证书名称_CCC/机动车辆及安全附件', '证书名称_CCC/音视频设备',
       '证书名称_CCC产品认证证书', '证书名称_其他自愿性工业产品认证', '有效期起止日期', '有效期截至日期', '状态_已撤销',
       '状态_已暂停', '状态_已注销', '状态_已过期', '状态_撤销', '状态_暂停', '状态_有效', '状态_正常',
       '状态_注销', '状态_过期失效', '资质认证_count']
```

```python
data=data.reindex(fillindex)
data.loc['状态_正常',:]=0
data
```
