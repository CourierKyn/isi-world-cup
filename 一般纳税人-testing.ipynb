```python
import pandas as pd
import numpy as np
import datetime
import dateutil
import re
```

```python
data=pd.read_excel('赛题1测试数据集/一般纳税人.xlsx',index_col='企业编号')
```

```python
data.columns
```

```python
data.drop_duplicates(keep='first',inplace=True)
data
```

```python
data_1=pd.get_dummies(data['纳税人资格'],prefix='纳税人资格')
```

```python
data_1=data_1.groupby('企业编号').sum().T
```

```python
data_1
```

```python
#规整日期格式
def convert(cell):
    pattern=r'\d+'
    if cell == '--' or len(re.findall(pattern,cell)) == 0:
        return np.nan
    return '-'.join(re.findall(pattern, cell))
```

```python
datatime=data[['认定日期']].applymap(str).applymap(convert)
datatime
```

```python
#计算时间间隔
def to_datetime(cell):
    if not isinstance(cell, str):
        return cell
    try:
        return (dateutil.parser.parse(cell) - datetime.datetime(2019, 3, 5)).days
    except:
        print(cell)
```

```python
data_2=datatime.applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_2
```

```python
data_3=data[['有效日期期起']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_3
```

```python
data_4=data[['有效截止日期']].applymap(to_datetime).groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526)))
data_4
```

```python
def limit(cell):
    if cell > 1:
        return 1
    return cell
```

```python
data_5=pd.get_dummies(data['登记注册类型'],prefix='登记注册类型').groupby('企业编号').sum().T.applymap(limit).reindex(columns=list(range(4001,4526))).fillna(0)
data_5
```

```python
data_6=pd.get_dummies(data['纳税人状态'],prefix='纳税人状态').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526))).fillna(0)
data_6
```

```python
data_7=pd.get_dummies(data['扣缴义务'],prefix='扣缴义务').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526))).fillna(0)
data_7
```

```python
data_8=pd.get_dummies(data['出口状态备案状态'],prefix='出口状态备案状态').groupby('企业编号').mean().T.reindex(columns=list(range(4001,4526))).fillna(0)
data_8
```
